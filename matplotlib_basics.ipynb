```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
```

```python
# 1st method
fig = plt.figure() # create figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method
fig, ax = plt.subplots()
ax.plot(x, y);
```

## Matplotlib example workflow

```python
# prepare data
# setup plot
# plot data
# customize plot
# save and show figure

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y)
ax.set(title="Simple plot", xlabel="x-axis", ylabel="y-axis")
fig.savefig("images/plot1_example.png")
```

### Plot from dataframe

```python
car_sales = pd.read_csv("data/car-sales.csv")
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2023", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]", "")
car_sales["Sale date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Sale date", y="Total sales")
```

```python
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# create a histogram of age
heart_disease["age"].plot.hist(bins=30);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
# pyplot method

over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
);

# customize
ax.set(title="heart disease and cholesterol levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target");

# add horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```
